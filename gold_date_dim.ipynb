# fetch relative columns

```python
df_source=spark.sql('''
select DISTINCT(Date_ID) AS Date_ID
from parquet.`abfss://silver@azurecarsalesdata7.dfs.core.windows.net/carsales`
''')
```

```python
display(df_source)
```

dim_date sink -Initial and incremental

```python
if spark.catalog.tableExists('cars_sales_catalog.silver.dim_date'):
   df_sink=spark.sql('''
   select dim_date_key,Date_ID
   from cars_sales_catalog.silver.dim_date
   ''')


else:
  df_sink=spark.sql('''
  select 1 as  dim_date_key,Date_ID
  from parquet.`abfss://silver@azurecarsalesdata7.dfs.core.windows.net/carsales`
  where 1=0''')#get only schema for initial run

```

```python
display(df_sink)
```

#existing records


```python
df_filter = df_source.join(df_sink,df_source['Date_ID']==df_sink['Date_ID'],'left').select(df_source['Date_ID'],df_sink['dim_date_key'])#left join source and sink 
```

```python
display(df_filter)
```

# Filtering new records and old records

```python
df_filter_old= df_filter.filter(col('dim_date_key').isNotNull())
```

```python
display(df_filter_old)#all are null
```

```python
df_filter_new= df_filter.filter(col('dim_date_key').isNull()).select(df_source['Date_ID'])#get dealer id,dealer name from where dim dealer key is null
```

```python
display(df_filter_new)
```

#create surrogate key

#fetch the max surrogate key from table

```python
if (incremental_flag=='0'):
    max_value=1#initial load
else:
    max_value_df=spark.sql("select max(dim_date_key) from cars_sales_catalog.silver.dim_date")#list
    max_value=max_value_df.collect()[0][0]+1#select the value from the list
```

#create surrogate key column and add max surrrogate key

```python
df_filter_new=df_filter_new.withColumn('dim_date_key',max_value+monotonically_increasing_id())#create a column dim_dealer_key which increments the max value
```

```python
display(df_filter_new)
```

#create final df

```python
df_final=df_filter_new.union(df_filter_old)
```

```python
display(df_final)
```

```python
from delta.tables import DeltaTable
```

#SCD TYPE 1

```python
if spark.catalog.tableExists('cars_sales_catalog.silver.dim_date'):
    delta_table=DeltaTable.forPath(spark,'abfss://gold@azurecarsalesdata7.dfs.core.windows.net/dim_date')
    delta_table.alias("target").merge(df_final.alias("source"),"target.dim_date_key=source.dim_date_key")\
                               .whenMatchedUpdateAll()\
                               .whenNotMatchedInsertAll()\
                               .execute()

else:#initial run
    df_final.write.format('delta')\
            .mode('overwrite')\
            .option('path',"abfss://gold@azurecarsalesdata7.dfs.core.windows.net/dim_date")\
            .saveAsTable('cars_sales_catalog.silver.dim_date')
```
